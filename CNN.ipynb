```python
# Importing the required libraries
import cartopy.crs as ccrs
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import xarray as xr
# import fsspec

from __future__ import absolute_import, division, print_function

import os

import tensorflow as tf
from tensorflow import keras
print("TensorFlow version is ", tf.__version__)

import numpy as np

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
# files paths
gcloud_bucket = "gs://satellite-tttr-hackathon"
satellite_path = "satellite"
weather_path = 'weather'

no2_files = gcloud_bucket+"/"+satellite_path+"/NO2/no2_kgm2_"
cl_files = gcloud_bucket+"/"+satellite_path+"/cloud_fraction/cloud_fraction_"

uwind_files = gcloud_bucket+"/"+weather_path+"/wind/uwind_ms_"
vwind_files = gcloud_bucket+"/"+weather_path+"/wind/vwind_ms_"
```

```python
%%time
import gcsfs

# Create a GCSFileSystem object
fs = gcsfs.GCSFileSystem()

# Use wild-card matching to get a list of file paths
no2_paths = fs.glob(f'{no2_files}2023092*.tiff')

# Now, you can use the list of file paths as needed
no2_data = []
for file_path in no2_paths:
    wth = xr.open_dataarray("gs://"+file_path).values.reshape(241, 162)
    no2_data.append(wth)

# Use wild-card matching to get a list of file paths
uwind_paths = fs.glob(f'{uwind_files}2023092*.tiff')
uwind_data = []
for file_path in uwind_paths:
    wth = xr.open_dataarray("gs://"+file_path).values.reshape(241, 162)
    uwind_data.append(wth)

# Use wild-card matching to get a list of file paths
vwind_paths = fs.glob(f'{vwind_files}2023092*.tiff')
vwind_data = []
for file_path in vwind_paths:
    wth = xr.open_dataarray("gs://"+file_path).values.reshape(241, 162)
    vwind_data.append(wth)
    
all_data = [np.stack([n, u, v], axis=2) for n, u, v in zip(no2_data, uwind_data, vwind_data)]
```

```python
labels = pd.read_csv(gcloud_bucket+'/Taiwan_nox_emissions.csv')
labels = labels[(labels['facility_id']=='C1400170')&
                (labels['datetime']>='2023-09-20')&
                (labels['datetime']<'2023-09-30')]
labels
```

```python
from keras.applications import ResNet50
from keras.models import Model
from keras.layers import Dense, GlobalAveragePooling2D
from keras.optimizers import Adam
from keras.preprocessing.image import ImageDataGenerator

image_size = all_data[0].shape

# Load the ResNet50 model without the top (fully connected) layers
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=all_data[0].shape)

# Freeze the layers of the base model
for layer in base_model.layers:
    layer.trainable = False

# Add your regression head
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(128, activation='relu')(x)
predictions = Dense(1, activation='linear')(x)

# Create the final model
model = Model(inputs=base_model.input, outputs=predictions)

# Compile the model
# model.compile(optimizer=Adam(), loss='mean_squared_error')  # Use appropriate loss for regression task
model.compile(loss='mean_squared_error')  # Use appropriate loss for regression task

# Print the model summary
model.summary()
```

```python
# x_train = np.array(all_data)
x_train = np.nan_to_num(all_data, nan=0)
y_train = np.array(labels.value)

# Create an ImageDataGenerator for data augmentation and normalization
datagen = ImageDataGenerator(validation_split=0)

# Training data generator
train_generator = datagen.flow(x_train, y_train, subset='training', 
                               ignore_class_split=True)

# Train the model
model.fit(train_generator, epochs=10)
```

```python
predictions = model.predict(x_train)

# Print the regression output
print("Regression Output:", predictions)
```
